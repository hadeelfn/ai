this a simple practical project. Finding anagrams in the dictionary of English words. As you know, two words are anagrams of each other when their letters can be rearranged to turn one word into the other.
 For instance, Elvis and lives are anagrams of each other. In this project.

```python
word = open('words.txt','r')
```

```python
word
```

```python
wordlist = word.readlines()

```

```python
wordlist[:10]
```

```python
len(wordlist)
```

```python
'Aaron\n'.strip()
```

```python
'Aaron'.lower()
```

```python
wordclean = [word.strip().lower() for word in wordlist]
```

```python
wordclean[:10]
```

```python
wordunique = list(set(wordclean))
```

```python
wordunique
```

```python
wordunique.sort()
```

```python
wordunique[:10]
```
