```python
wordclean = [word.strip().lower() for word in open('words.txt','r')]
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words.txt','r')]))) #اختصار لكل الخطوات الي قبل 
```

```python
wordclean[:10]
```

```python
len(wordclean)
```

```python
def signature(word): #Finding Anagrams
    return ''.join(sorted(word))
```

```python
def anagram(myword):
    return [word for word in wordclean if signature(word) == signature(myword)]
```

```python
anagram('dictionary')    #test 
```

*This solution however is very costly

```python
import collections
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
def anagram_fast(myword):
    return words_bysig[signature(myword)]
```

```python
anagram_fast('dictionary')
```

```python
anagrams = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```
